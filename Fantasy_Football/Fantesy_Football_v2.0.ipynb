```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://www.pro-football-reference.com/years/2022/scrimmage.htm"
df = pd.read_html(url)[0]

print(df.head())

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table')

print(table)

links = []
for tr in table.findAll("tr"):
    
    trs = tr.findAll("td")
    for each in trs:
        try:
            link = each.find('a')['href']
            if link.find('teams') == -1:
                links.append(link)
        except:
            pass
        
print(links)

```

```python
for n, name in enumerate(df['Unnamed: 1_level_0']['Player']):
    if name == 'Player':
        df = df.drop(n)
        
df = df.reset_index(drop=True)
```

```python
data = {'Name':[], 'Points':[], 'Points_Year':[]}
url = "https://www.pro-football-reference.com" + links[0]
data_df = pd.read_html(url)[0]
for n, m in data_df.columns:
    label = n + m
    data[label] = []
print(data)
```

```python
def get_Data(df, name):
    Year_List = df['Unnamed: 0_level_0']['Year'].to_list()

    for n, year in enumerate(Year_List):
        
        if isinstance(year, float):
            break
        year = year.replace('*', '', 1)
        year = year.replace('+', '', 1)
        Year_List[n] = year
        print(year)

    index = Year_List.index('2022')
    print(index)
    
    for n in range(index):
        data['Name'].append(name)
        
        if('Unnamed: 29_level_0' not in df.columns):
            break
        
        if(type(df['Unnamed: 29_level_0']['RRTD'][n  + 1] * 6) == str):
            break
        
        
        else:
            data['Points'].append(float(df['Unnamed: 29_level_0']['RRTD'][n  + 1] * 6) - float(df['Unnamed: 30_level_0']['Fmb'][n  + 1] * 2)
                    + float(df['Total Yds']['YScm'][n  + 1] * 0.3) + float(df['Receiving']['Rec'][n  + 1] * 1))
                              
            data['Points_Year'].append(Year_List[n + 1])
        
        for column, header in df.columns:
            label = column + header
            data[label].append(df[column][header].to_list()[n])
```

```python
import time
for n, link in enumerate(links):
    print(n)

    url = "https://www.pro-football-reference.com" + link
    print(url)
    player_df = pd.read_html(url)[0]
    name = df["Unnamed: 1_level_0"]["Player"][n]
    name = name.replace('*', '', 1)
    name = name.replace('+', '', 1)
    print(name)
    print(len(player_df))
    
    position = df["Unnamed: 4_level_0"]["Pos"][n]
    time.sleep(5)
    if name != 'Ray-Ray McCloud' and len(player_df) > 3 and (position == 'RB' or position == 'WR' or position == 'TE'):
        print(position)
        get_Data(player_df, name)
```
